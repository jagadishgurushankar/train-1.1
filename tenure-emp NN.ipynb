```python
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, RMSprop, SGD

def build_model(hp):
    model = Sequential()

    # tune activation
    activation = hp.Choice('activation', ['relu', 'tanh'])

    # tune optimizer
    optimizer = hp.Choice('optimizer', ['adam', 'rmsprop', 'sgd'])

    model.add(Dense(16, activation=activation, input_shape=(9,)))
    model.add(Dense(16, activation=activation))
    model.add(Dense(1, activation='sigmoid'))

    # pick optimizer
    if optimizer == "adam":
        opt = Adam()
    elif optimizer == "rmsprop":
        opt = RMSprop()
    else:
        opt = SGD()

    model.compile(
        optimizer=opt,
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model


tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,
)

tuner.search(x_train, y_train, epochs=50, validation_split=0.2)

best_model = tuner.get_best_models(1)[0]
best_model.summary()
```

## you can also include total number of nodes in the tuner

```python
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, RMSprop, SGD

def build_model(hp):
    model = Sequential()

    # tune activation
    activation = hp.Choice('activation', ['relu', 'tanh',"leaky_relu"])

    # tune optimizer
    optimizer = hp.Choice('optimizer', ['adam', 'rmsprop', 'sgd'])

    # tune number of nodes in hidden layers
    nodes = hp.Int('nodes', min_value=8, max_value=64, step=8)

    # hidden layers with tunable nodes
    model.add(Dense(nodes, activation=activation, input_shape=(9,)))
    model.add(Dense(nodes, activation=activation))

    # output layer
    model.add(Dense(1, activation='sigmoid'))

    # choose optimizer object
    if optimizer == "adam":
        opt = Adam()
    elif optimizer == "rmsprop":
        opt = RMSprop()
    else:
        opt = SGD()

    model.compile(
        optimizer=opt,
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model


tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,
)

tuner.search(x_train, y_train, epochs=50, validation_split=0.2)

best_model = tuner.get_best_models(1)[0]
best_model.summary()
```

```python
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, RMSprop, SGD,Adagrad

def build_model(hp):
    model = Sequential()

    # tune activation
    activation = hp.Choice('activation', ['relu', 'tanh',"leaky_relu"])

    # tune optimizer
    optimizer = hp.Choice('optimizer', ['adam', 'rmsprop', 'sgd'])

    # tune number of nodes in hidden layers
    nodes = hp.Int('nodes', min_value=8, max_value=64, step=8)

    # hidden layers with tunable nodes
    model.add(Dense(nodes, activation=activation, input_shape=(21,)))
    model.add(Dense(nodes, activation=activation))

    # output layer
    model.add(Dense(1, activation='linear'))

    # choose optimizer object
    if optimizer == "adam":
        opt = Adam()
    elif optimizer == "rmsprop":
        opt = RMSprop()
    else:
        opt = SGD()

    model.compile(
        optimizer=opt,
        loss='mse',
        metrics=['mae']
    )

    return model


tuner = kt.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=10,
)

tuner.search(x_train, y_train, epochs=50, validation_split=0.2)

best_model = tuner.get_best_models(1)[0]
best_model.summary()
```

	•	loss='mse' tells the model:

“While training, try to minimize the mean squared error between predictions and true values.”

	•	This affects the weights and biases of the network:
Every forward pass → compute output → compare with true value → compute MSE → backpropagate → update weights.

Where in the architecture?
	•	Applied at the output layer (last layer)
	•	Influences all layers indirectly via backpropagation.


 2. Metrics (MAE) – Only for monitoring
	•	metrics=['mae'] does not affect training.
	•	It just calculates the mean absolute error at the output layer after each epoch, so you can see how well the network is doing.

Where in the architecture?
	•	Only applied at the output predictions, after the forward pass.
	•	It does not feed into weight updates.
